```python
## Imports
import bs4 as bs
import urllib.request
import pandas as pd

# Sauce is the standard name for the ingredient to soup...get it?
sauce = urllib.request.urlopen('https://pythonprogramming.net/parsememcparseface/').read()
soup = bs.BeautifulSoup(sauce,'lxml')

# Now that we make the request and cook the soup, let's print a test...
print(soup.title.string)
```

```python
## Cool, so now let's take a look at all the pargraph text on the page like this
for paragraph in soup.find_all('p'):
    print(paragraph.text)
```

```python
# Or, get all the text!

print(soup.get_text())
```

___
Great, so now we have the ability to select different parts of the web page by using bs!
___

```python
# Okay, now let's take a look at the navigation on the site..

nav = soup.nav
for url in nav.find_all('a'):
    print(url.get('href'))
    
#He explains that the website really has two navbars, depending on the size of the screen one is chosen to be selectable.
```

___
So we've scraped the nav tags, what about div tags??
___

```python
# the find_all() passes parameters like this: find_all('text-to-find', class_='specific-class-tag')
# Let's give it a shot..

for div in soup.find_all('div', class_='body'):
    print(div.text)
```

___
Okay! We even got some of the text data in the div class='body' tags..cool. I think there will be better ways to parse the tables though. Just a hunch..
___

```python
# Implementnig table scraping in bs

table = soup.table
table_rows = table.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print (row)
```

___
This is all well and good, but apparently pandas is much better at parsing tables from sites. If someone is nice enough to place tables in the html of the website. You should probably use the pandas functionality. Let me show you...
___

```python
# Scraping tables with Pandas

dfs = pd.read_html('https://pythonprogramming.net/parsememcparseface/', header=0)
for df in dfs:
    print(df)
```
